User defined Functions
---

```python
salary_1 = 100000
rate_1 = 0.1

# total_salary = base_salary * (1 + bonus_rate)
# total_salary


# With function now!
def calculate_salary(base_salary, bonus_rate):

    total_salary = base_salary * (1 + bonus_rate)

    return total_salary

```

```python
calculate_salary(salary_1, rate_1)
```

Data filtering example
---

```python
jobs_data = [
    {'job_title': 'Data Analyst', 'job_skills': ['Python', 'SQL', 'Excel'], 'remote': True},
    {'job_title': 'Junior Data Analyst', 'job_skills': ['Python', 'SQL', 'Excel'], 'remote': True},
    {'job_title': 'Data Scientist', 'job_skills': ['Python', 'R', 'Machine Learning', 'Deep Learning'], 'remote': False},
    {'job_title': 'Machine Learning Engineer', 'job_skills': ['Python', 'TensorFlow', 'PyTorch', 'Scikit-Learn'], 'remote': False},
    {'job_title': 'Data Engineer', 'job_skills': ['Apache Spark', 'Hadoop', 'SQL'], 'remote': True}
]
```

Finding remote jobs True

```python
list(filter(lambda job: job['remote'] and "Python" in job['job_skills'], jobs_data))
```
